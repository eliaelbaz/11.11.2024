```python
# 1
import pandas as pd

games = [
    "Planet Coaster 2", "Mario & Luigi: Brothership", "Slitterhead",
    "The Rise of the Golden Idol", "Dragon Quest 3 HD-2D Remake",
    "Microsoft Flight Simulator 2024", "Stalker 2: Heart of Chornobyl",
    "Fantasian Neo Dimension", "Marvel Rivals",
    "Indiana Jones and the Great Circle", "Blood Line: A Rebel Moon Game",
    "The Legend of Zelda: Echoes of Wisdom"
]

# a
print("Games List:")
print(games)

# b
print("\nGame at index 3:")
print(games[3])

# c
prices = [
    224.96, 187.46, 149.96, 131.21, 187.46, 262.46, 224.96,
    168.71, 149.96, 206.21, 0.0, 262.46
]
game_prices = pd.Series(prices, index=games)
print("\nGame Prices Series:")
print(game_prices)

# d
print("\nGame Prices Series:")
print(game_prices)

# e
print("\nPrice of Marvel Rivals:")
print(game_prices["Marvel Rivals"])

# f
print("\nKeys of the series:")
print(game_prices.index)

# g
print("\nData type of the series:")
print(game_prices.dtype)

# h
selected_games = {
    "Planet Coaster 2": 224.96,
    "Slitterhead": 149.96,
    "Marvel Rivals": 149.96
}
selected_series = pd.Series(selected_games)
print("\nSelected Games Series:")
print(selected_series)

# i
reset_series = selected_series.reset_index()
print("\nSeries with reset index:")
print(reset_series)
```

```python
# 2
# + מבצע חיבור ישיר על שני ערכים באותו אינדקס, ואם חסר ערך הוא יחזיר NaN.
# add מאפשר להוסיף ערך ברירת מחדל (fill_value) במידה וחסר ערך באחד הצדדים
s1 = pd.Series([1, 2, 3], index=["a", "b", "c"])
s2 = pd.Series([4, 5], index=["a", "b"])

print(s1 + s2)
print(s1.add(s2, fill_value=0))
```

```python
# 3
game_prices = pd.Series({
    "Planet Coaster 2": 224.96,
    "Mario & Luigi: Brothership": 187.46,
    "Slitterhead": 149.96,
    "Marvel Rivals": 149.96,
    "The Legend of Zelda: Echoes of Wisdom": 262.46
})

# הנחות
discounts = pd.Series({
    "Planet Coaster 2": 50,
    "Marvel Rivals": 80,
    "The Legend of Zelda: Echoes of Wisdom": 15
})

# חישוב מחירים חדשים
new_prices = game_prices.subtract(discounts, fill_value=0)
print("\nNew Prices after Discounts:")
print(new_prices)
```
